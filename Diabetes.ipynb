```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer([('oh', OneHotEncoder(), [0])], remainder='passthrough')
y = ct.fit_transform(y)
```

# Splitting dataset to train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# Feature Engineering

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train[0]
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

model = Sequential()
```

```python
model.add(Dense(units=8, kernel_initializer='random_uniform', activation='sigmoid'))  # input layer
model.add(Dense(units=16, kernel_initializer='random_uniform', activation='sigmoid')) # hidden layer
model.add(Dense(units=2, kernel_initializer='random_uniform', activation='sigmoid'))  # output layer
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy']) # optimizer
```

```python
model.fit(x_train, y_train, epochs=7000)
```

```python
ypred = model.predict(x_test)
```
